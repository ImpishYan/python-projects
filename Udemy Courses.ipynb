```python
#Find out number of courses per subject
data['subject'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='subject', data=data)
plt.xlabel('Subjects',fontsize=13)
plt.ylabel('Number of Courses per Subject')
plt.xticks(rotation=60)
```

```python
data['level'].value_counts()
```

```python
sns.countplot(x='level', data=data)
plt.xlabel('Levels',fontsize=13)
plt.ylabel('Number of Courses per Level')
plt.xticks(rotation=60)
```

```python
data['is_paid'].value_counts()
```

```python
sns.countplot(x='is_paid', data=data)
plt.xlabel('Paid',fontsize=13)
plt.ylabel('Number of Free and Paid Courses')
plt.xticks(rotation=60)
```

```python
#which course has more lectures?
data.groupby(['is_paid']).mean()['num_lectures']
```

```python
sns.barplot(x='is_paid', y='num_subscribers',data=data)
```

```python
#Which level has the highest number of subscribers?
```

```python
sns.barplot(x='level',y='num_subscribers',data=data)
plt.xticks(rotation=30)
plt.show()
```

```python
#Find the most popular course title
```

```python
data[data['num_subscribers'].max()==data['num_subscribers']]['course_title']
```

```python
#Display 10 most popular course based on subscribers
```

```python
top_10=data.sort_values(by='num_subscribers',ascending=False).head(10)
```

```python
sns.barplot(x='num_subscribers',y='course_title',data=top_10)
```

```python
#Find course with the highest number of reviews
sns.barplot(x='subject',y='num_reviews',data=data)
plt.xticks(rotation=30)
plt.show()
```

```python
#Does price affect number of reviews?
```

```python
plt.figure(figsize=(15,3))
sns.scatterplot(x='price',y='num_reviews',data=data)
```
